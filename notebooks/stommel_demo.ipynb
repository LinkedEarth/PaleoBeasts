# Stommel Model Demo

This notebook demonstrates the new `Stommel` model in `paleobeasts.signal_models`.

```python
import numpy as np
import matplotlib.pyplot as plt
import paleobeasts as pb
from paleobeasts.signal_models import Stommel
```

```python
t_span = (0, 40)
y0 = [1.0, 0.2]

model_unforced = Stommel(forcing=None, alpha=1.0, beta=1.0, k=1.2, E=0.0)
model_unforced.integrate(t_span=t_span, y0=y0, method='RK45')

forcing = pb.Forcing(lambda t: 0.15 * np.sin(2 * np.pi * t / 10.0))
model_forced = Stommel(forcing=forcing, alpha=1.0, beta=1.0, k=1.2, E=0.0)
model_forced.integrate(t_span=t_span, y0=y0, method='RK45')
```

```python
fig, ax = plt.subplots(3, 1, figsize=(9, 8), sharex=True)

ax[0].plot(model_unforced.time, model_unforced.state_variables['T'], label='Unforced')
ax[0].plot(model_forced.time, model_forced.state_variables['T'], label='Forced', alpha=0.9)
ax[0].set_ylabel('T')
ax[0].legend()

ax[1].plot(model_unforced.time, model_unforced.state_variables['S'], label='Unforced')
ax[1].plot(model_forced.time, model_forced.state_variables['S'], label='Forced', alpha=0.9)
ax[1].set_ylabel('S')
ax[1].legend()

ax[2].plot(model_unforced.time, model_unforced.diagnostic_variables['q'], label='Unforced q')
ax[2].plot(model_forced.time, model_forced.diagnostic_variables['q'], label='Forced q', alpha=0.9)
ax[2].set_ylabel('q')
ax[2].set_xlabel('Time')
ax[2].legend()

plt.tight_layout()
plt.show()
```
